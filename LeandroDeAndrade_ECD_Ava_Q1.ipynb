```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

observacoes = np.array([8, 10, 5, 6, 8, 9, 6, 11, 9, 8, 7, 7])
mu_obs = observacoes.mean()
s = np.std(observacoes, ddof=1)
n = len(observacoes)
mu_h0 = 8
sig = 0.05
df = n - 1

t_calc = (mu_obs - mu_h0) / (s / np.sqrt(n))

t_crit = scipy.stats.t.ppf(1 - sig, df)

p_value = scipy.stats.t.sf(t_calc, df)

print(f"Estatística t calculada: {t_calc:.3f}")
print(f"Valor t crítico (alfa={sig}, unicaudal direita): {t_crit:.3f}")
print(f"Valor p: {p_value:.4f}")

x = np.linspace(scipy.stats.t.ppf(0.001, df), scipy.stats.t.ppf(0.999, df), 100)
pdf = scipy.stats.t.pdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'k-', lw=2, label=f'Distribuição t (df={df})')

plt.axvline(t_calc, color='blue', linestyle='--', label=f't Calculado ({t_calc:.3f})')

plt.axvline(t_crit, color='red', linestyle='--', label=f't Crítico ({t_crit:.3f}) - α={sig}')

x_rej = np.linspace(t_crit, scipy.stats.t.ppf(0.999, df), 100)
plt.fill_between(x_rej, scipy.stats.t.pdf(x_rej, df), color='red', alpha=0.3, label='Região de Rejeição')

plt.xlabel('Valores t')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição t e Resultado do Teste de Hipótese (Unicaudal Direita)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.text(0, plt.ylim()[1]*0.9, f'Valor p = {p_value:.4f}', color='green', fontsize=12)

plt.show()
```
